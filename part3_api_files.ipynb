```python
print("="*50)
print("Task 1 — File Read & Write Basics")
print("="*50)


lines = [
    "Topic 1: Variables store data. Python is dynamically typed.\n",
    "Topic 2: Lists are ordered and mutable.\n",
    "Topic 3: Dictionaries store key-value pairs.\n",
    "Topic 4: Loops automate repetitive tasks.\n",
    "Topic 5: Exception handling prevents crashes.\n"
]
# Part 3 - Task1

# Part A — Write

with open("python_notes.txt", "w", encoding="utf-8") as file:
    file.writelines(lines)

print("File written successfully.")

with open("python_notes.txt", "a", encoding="utf-8") as file:
    file.write("Topic 6: Functions help reuse code.\n")
    file.write("Topic 7: File handling allows persistent storage.\n")

print("Lines appended successfully.")


# Part B — Read

with open("python_notes.txt", "r", encoding="utf-8") as file:
    content = file.readlines()

print("\nFile Content:")

for i, line in enumerate(content, start=1):
    print(f"{i}. {line.strip()}")

print("\nTotal number of lines:", len(content))


# Keyword search

keyword = input("\nEnter keyword to search: ").lower()

found = False

print("\nMatching lines:")

for line in content:
    if keyword in line.lower():
        print(line.strip())
        found = True

if not found:
    print("No matching lines found.")

########################################################################################


# Part 3 - Task2

print("="*87)
print("Task 2 — API Integration")
print("="*87)


import requests

# Step 1 — Fetch and Display Products

url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)

data = response.json()
products = data["products"]

print("Product List:")
print(f"{'ID':<4} | {'Title':<43} | {'Category':<12} | {'Price':<8} | {'Rating'}")
print("-"*87)

for p in products:
    print(f"{p['id']:<4} | {p['title']:<43} | {p['category']:<12} | ${p['price']:<7} | {p['rating']}")


# Step 2 — Filter and Sort

filtered = []

for p in products:
    if p["rating"] >= 4.5:
        filtered.append(p)

filtered.sort(key=lambda x: x["price"], reverse=True)

print("Filtered (Rating ≥ 4.5) & Sorted by Price:")
print("-"*87)

for p in filtered:
    print(f"{p['title']} | ${p['price']} | Rating: {p['rating']}")


# Step 3 — Search by Category (Laptops)

laptop_url = "https://dummyjson.com/products/category/laptops"
response = requests.get(laptop_url)

laptops = response.json()["products"]

print("Laptop Products:")
print("-"*87)

for p in laptops:
    print(f"{p['title']} — ${p['price']}")


# Step 4 — POST Request

post_url = "https://dummyjson.com/products/add"

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

response = requests.post(post_url, json=new_product)

print("POST Response:")
print(response.json())


########################################################################################


# Part 3 - Task3

print("="*50)
print("Task 3 — Exception Handling")
print("="*50)

import requests


# Part A — Guarded Calculator

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("Part A — Safe Divide:")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# Part B — Guarded File Reader

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

print("Part B — File Reader:")
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))


# Part C — Robust API Calls

print("Part C — API Calls:")

try:
    response = requests.get("https://dummyjson.com/products?limit=5", timeout=5)
    data = response.json()
    print("Fetched products successfully.")
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print("Error:", e)


try:
    response = requests.post(
        "https://dummyjson.com/products/add",
        json={"title": "Test Product", "price": 100},
        timeout=5
    )
    print("POST request successful:", response.json())
except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print("Error:", e)


# Part D — Input Validation Loop

print("Part D — Product Lookup:")

while True:
    user_input = input("Enter product ID (1–100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting...")
        break

    if not user_input.isdigit():
        print("Invalid input! Please enter a number.")
        continue

    product_id = int(user_input)

    if product_id < 1 or product_id > 100:
        print("Please enter a number between 1 and 100.")
        continue

    try:
        url = f"https://dummyjson.com/products/{product_id}"
        response = requests.get(url, timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            product = response.json()
            print(f"{product['title']} — ₹{product['price']}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Error:", e)


########################################################################################


# Part 3 - Task4


print("="*50)
print("Task 4 — Logging to File")
print("="*50)

import requests
from datetime import datetime


# Logger function

def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"

    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_entry)


# Trigger 1 — ConnectionError

try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    print("Connection error occurred.")
    log_error("fetch_products", "ConnectionError", str(e))


# Trigger 2 — HTTP 404 (NOT exception)

try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)

    if response.status_code != 200:
        message = f"{response.status_code} Not Found for product ID 999"
        print("HTTP error occurred.")
        log_error("lookup_product", "HTTPError", message)

except requests.exceptions.ConnectionError as e:
    log_error("lookup_product", "ConnectionError", str(e))
except requests.exceptions.Timeout as e:
    log_error("lookup_product", "Timeout", str(e))
except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))


# Read and print log file

print("\nError Log Contents:")
print("-"*50)

try:
    with open("error_log.txt", "r", encoding="utf-8") as file:
        print(file.read())
except FileNotFoundError:
    print("No logs found.")
```
